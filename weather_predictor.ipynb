```python
weather["target"] = weather.shift(-1)["tmax"]
```

```python
weather = weather.ffill()
weather.columns
```

```python
from sklearn.linear_model import Ridge
rr = Ridge(alpha=.1)
```

```python
predictors = weather.columns[~weather.columns.isin(["target", "name", "station"])]
predictors
```

```python

def backtest(weather, model, predictors, start=3650, step=90):
    all_predictions = []
    for i in range(start, weather.shape[0], step):
        train = weather.iloc[:i,:]
        test = weather.iloc[i:(i+step),:]
        model.fit(train[predictors], train["target"])
        
        preds = model.predict(test[predictors])
        preds = pd.Series(preds, index=test.index)
        combined = pd.concat([test["target"], preds], axis=1)
        combined.columns = ["actual", "prediction"]
        combined["diff"] = (combined["prediction"] - combined["actual"]).abs()
        
        all_predictions.append(combined)
        
    return pd.concat(all_predictions)
predictions = backtest(weather, rr, predictors)
predictions['prediction']=predictions['prediction'].round(2).abs()
predictions['diff']=predictions['prediction'].round(2).abs()
predictions['actual']=predictions['actual'].round(2).abs()
predictions.index
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
mean_absolute_error(predictions["actual"], predictions["prediction"])
```

```python
from sklearn.metrics import mean_absolute_error

mae = mean_absolute_error(predictions["actual"], predictions["prediction"])
print("Mean Absolute Error:", mae)
mean_squared_error(predictions["actual"], predictions["prediction"])
```

```python
print('ENTER A SPECIFIC MONTH AND DATE (MM-DD) TO GET 10-YEAR AVERAGE DATA:')
month_day = input()

predictions['month_day'] = predictions.index.strftime('%m-%d')
result_df = predictions.groupby('month_day')[['actual', 'prediction', 'diff']].mean().round(2)
result_df['sort_date'] = pd.to_datetime('2000-' + result_df.index)
result_df = result_df.sort_values('sort_date').drop(columns='sort_date').reset_index()
filtered = result_df[result_df['month_day'] == month_day]

if not filtered.empty:
    print(f"\n10-Year Average for {month_day}:")
    print(filtered[['prediction']])
else:
    print("\nNo data found for the entered month-day. Use format like '02-02'.")
```
